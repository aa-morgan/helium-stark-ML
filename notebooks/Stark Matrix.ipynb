```python
# packages
import os
import numpy as np
from hsML import Hamiltonian, h, c
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# construct matrix
nmin = 36
nmax = 36
S = 1
ML = None
mat = Hamiltonian(nmin, nmax, S=S, ML=ML)
print('Number of basis states:', '%d'%mat.num_states)
mat.basis
```

```python
# Stark matrix
sm = mat.stark_matrix()
# flatten matrix and sort
vals = np.triu(sm).flatten()
mask = vals != 0
stark_el = np.sort(vals[mask])
```

```python
fig, ax = plt.subplots(2, 1, figsize=(6, 8))

ax[0].plot(stark_el)
ax[0].set_xlabel('index')
ax[0].set_ylabel(r"$\langle 36, L', 1, J', 0 | \hat{H}_{s} | 36, L, 1, J, 0 \rangle$")

bins=np.arange(-850, 1000, 100)
ax[1].hist(stark_el, bins=bins, orientation='horizontal')

ax[1].set_xlabel('counts')
ax[1].set_ylabel(r"$\langle 36, L', 1, J', 0 | \hat{H}_{s} | 36, L, 1, J, 0 \rangle$")

plt.tight_layout()
# output
fileout = os.path.join("..", "images", "stark_histogram.png")
plt.savefig(fileout, dpi=200)
plt.show()
```
